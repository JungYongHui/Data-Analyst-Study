### 전체 이적료의 histogram 확인

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
# train_df['value'].hist()
plt.boxplot(train_df['value'])
```

```python
train_df['value'].describe()
```

```python
Q1 = train_df['value'].describe()['25%']
Q3 = train_df['value'].describe()['75%']
IQR = Q3-Q1

# 이상치 데이터 확인
# train_df[(train_df['value'] < Q1 - 1.5*IQR)or(train_df['value'] > Q3 + 1.5*IQR)]
temp = train_df.loc[(train_df['value'] > (Q1 - 1.5*IQR)) & (train_df['value'] < (Q3 + 1.5*IQR)),:]
plt.figure(figsize=(12,8))
plt.hist(temp['value'])
plt.show()
train_df.loc[(train_df['value'] < (Q1 - 1.5*IQR)) | (train_df['value'] > (Q3 + 1.5*IQR)),:]
```

- 일반적인 이상치 분류의 기준인 Q1 - 1.5*IQR, Q3 + 1.5*IQR 를 기준으로 하기에는 전체 8932 row 중 1178 row 가 이상치로 판단이 된다.
