```python
def datetime(date):
    new_date1 = date.replace('년 ', '-')
    new_date2 = new_date1.replace('월 ', '-')
    new_date3 = new_date2.replace('일', '')
    return new_date3

def price(price):
    new_price = price.replace(',', '')
    return new_price
```

```python
df_temp = df.copy()
df_temp['날짜'] = df['날짜'].apply(datetime)
feature = ['종가', '오픈', '고가', '저가']
for i in feature:
    df_temp[i] = df[i].apply(price)
df_temp['거래량'] = df_temp['거래량'].apply(lambda x:x[:-1])
df_temp['변동 %'] = df_temp['변동 %'].apply(lambda x:x[:-1])
df_temp.columns = ['date', 'close', 'open', 'high', 'low', 'volume K', 'rate %']
df_temp
```

```python
# z-test로 p-value 확인

from statsmodels.stats.weightstats import ztest

_, p = ztest(df.close)
print(p)  # 0
```

- p value 0.05 이하이므로 정규분포와는 거리가 멀다는 의미로 해석할 수 있다.
- 즉, 이 데이터로 정규분포를 가정한 신뢰구간 분석은 적용하기 어렵

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# 단기이평선 값인 5일을 기준으로 도출
# extrapolate_trend='freq' : Trend 성분을 만들기 위한 rolling window 때문에 필연적으로 trend, resid에는 Nan 값이 발생하기 때문에, 이 NaN값을 채워주는 옵션이다.
result = seasonal_decompose(df.close, model='additive', two_sided=True, 
                            period=5, extrapolate_trend='freq') 
plt.figure(figsize=(15,10))
result.plot()
plt.show()
```

### 시계열 분해

- additive 방법
    - Observed[t] = trend[t] + seasonal[t] + resid[t]
- multiplicative 방법
    - Observed[t] = trend[t] x seasonal[t] x resid[t]  
      위 식의 양변에 로그를 취하면 결과적으로 additive와 동일한 결과가 얻어진다.
      
> 잔차는 평균 0을 기준으로 분포하고 있으므로, 잔차가 큰 날은 일반적인 추세나 게절성에서 벗어난 날이다.

```python
fig, ax = plt.subplots(figsize=(9,6))
_ = plt.hist(result.resid, 100, density=True, alpha=0.75)
```

```python
r = result.resid.values
st, p = ztest(r)
print(st,p)  # 0.13945750234754367 0.8890886359287032
```

- p-value가 0.05 이상으로 나오므로, 잔차는 정규분포를 잘 따르고 있다고 판단

### 잔차(residual)가 큰 날짜의 데이터 확인 (이상치)

```python
# 평균과 표준편차 출력
mu, std = result.resid.mean(), result.resid.std()
print("평균:", mu, "표준편차:", std)
# 평균: 1.9089768239690312 표준편차: 611.2534875222074

# 3-sigma(표준편차)를 기준으로 이상치 판단
print("이상치 갯수:", len(result.resid[(result.resid>mu+3*std)|(result.resid<mu-3*std)]))
```

```python
print([i for i in df.date[result.resid[(result.resid>mu+3*std)|(result.resid<mu-3*std)].index]])
```

```python
def my_decompose(df, features, freq=5):
    trend = pd.DataFrame()
    seasonal = pd.DataFrame()
    resid = pd.DataFrame()
    
    for f in features:
        result = seasonal_decompose(df[f], model='additive', period=freq, extrapolate_trend=freq)
        trend[f] = result.trend.values
        seasonal[f] = result.seasonal.values
        resid[f] = result.resid.values
        
    return trend, seasonal, resid

# 각 변수별 트렌드/계절적/잔차
tdf5, sdf5, rdf5 = my_decompose(df, features=['open','high','low','close','volume K', 'rate %'], freq=5)
tdf20, sdf20, rdf20 = my_decompose(df, features=['open','high','low','close','volume K', 'rate %'], freq=20)
tdf60, sdf60, rdf60 = my_decompose(df, features=['open','high','low','close','volume K', 'rate %'], freq=60)
tdf120, sdf120, rdf120 = my_decompose(df, features=['open','high','low','close','volume K', 'rate %'], freq=120)
tdf5.describe()

fig = plt.figure(figsize=(15,7))
ax = fig.add_subplot(111)
df.close.plot(ax=ax, label='Observed', legend=True, color='k', linewidth=2)
tdf5.close.plot(ax=ax, label='Trend 5', legend=True, color='r', linewidth=1)
tdf20.close.plot(ax=ax, label='Trend 20', legend=True, color='yellow', alpha=0.5, linewidth=1)
tdf60.close.plot(ax=ax, label='Trend 60', legend=True, color='cornflowerblue', linewidth=1)
tdf120.close.plot(ax=ax, label='Trend 120', legend=True, color='green', linewidth=1)
rdf5.close.plot(ax=ax,label='Resid', legend=True)
plt.show()
```
